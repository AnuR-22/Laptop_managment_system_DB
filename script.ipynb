# [dbo].[usp_AddLaptop]

```sql
/****** Object:  StoredProcedure [dbo].[usp_AddLaptop]    Script Date: 09-06-2026 14:56:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[usp_AddLaptop]
    @Code NVARCHAR(50),
    @Brand NVARCHAR(50),
    @Model NVARCHAR(50),
    @SerialNumber NVARCHAR(100),
    @UserName NVARCHAR(100),
    @Department NVARCHAR(100),
    @IssueDate DATE,
    @Condition NVARCHAR(20),
    @ImagePath NVARCHAR(500)
AS
BEGIN
    INSERT INTO Laptops
    (
        Code, Brand, Model, SerialNumber,
        UserName, Department, IssueDate,
        Condition, ImagePath
    )
    VALUES
    (
        @Code, @Brand, @Model, @SerialNumber,
        @UserName, @Department, @IssueDate,
        @Condition, @ImagePath
    );
END
GO
```

# [dbo].[usp_GetLaptopByCode]

```sql
/****** Object:  StoredProcedure [dbo].[usp_GetLaptopByCode]    Script Date: 09-06-2026 14:56:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[usp_GetLaptopByCode]
    @Code NVARCHAR(50)
AS
BEGIN
    SELECT 
        Code,
        Brand,
        Model,
        SerialNumber,
        UserName,
        Department,
        IssueDate,
        Condition,
        ImagePath
    FROM Laptops
    WHERE Code = @Code;
END
GO
```
